```python
from scipy.signal import chirp, spectrogram
import matplotlib.pyplot as plt
import numpy as np

fs = 8000
T = 10
t = np.linspace(0, T, T*fs, endpoint=False)
w = chirp(t, f0=1500, f1=250, t1=10,phi=0, method='quadratic')

ff, tt, Sxx = spectrogram(w, fs=fs, noverlap=256, nperseg=512,
                          nfft=2048)


plt.pcolormesh(tt, ff[:513], Sxx[:513], cmap='gray_r')
plt.title('Quadratic Chirp, f(0)=1500, f(10)=250')
plt.xlabel('t (sec)')
plt.ylabel('Frequency (Hz)')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
# import plotly.plotly as py
import numpy as np
import pandas as pd


Fs = 1024.0;  # sampling rate
Ts = 1.0/Fs; # sampling interval
fft_size = Fs/2

t = np.arange(0,1,Ts) # time vector

ff = 200;   # frequency of the signal


y = np.sin(2*np.pi*ff*t)
y = y+ 0.2*np.random.normal(0.0,1.0,len(y))

n = len(y) # length of the signal

print(n)
k = np.arange(n)
T = n/Fs
frq = k/T # two sides frequency range


print(range(int(n/2)))
frq = frq[range(int(n/2))] # one side frequency rangen

Y = np.fft.fft(y)/fft_size # fft computing and normalization
Y = Y[range(int(n/2))]

fig, ax= plt.subplots(2, 1)
fig.subplots_adjust(hspace=0.6)

ax[0].plot(t[:100],y[:100])
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Amplitude')

ax[1].plot(frq/1000,abs(Y),'r') # plotting the spectrum
ax[1].set_xlabel('Freq (KHz)')
ax[1].set_ylabel('|Y(freq)|')

plt.show()

# plot_url = py.plot_mpl(fig, filename='mpl-basic-fft')
```
