```python
filename = 'data/YemenCholera.csv'
table = read_csv(filename, header=0, index_col=0, decimal='M', encoding = "utf-8")
table.columns = ['Gov','Cases', 'Deaths', 'CFR', 'AR', 'PCode']
```

```python
Gov = table.Gov
Cases = table.Cases
Deaths = table.Deaths
CFR = table.CFR
AR = table.AR
PCode = table.PCode;
```

```python
counter = 0
index = Cases.index[0]
values = [0]
for i in range(len(Cases)):
    if (Cases.index[i] == index):
        values[counter] += int(''.join([i for i in Cases[i] if i != ',']))
    else:
        counter += 1
        index = Cases.index[i]
        values.append(0)
        values[counter] += int(''.join([i for i in Cases[i] if i != ',']))
print(values)
```

```python
plot(weeks, values, 'r', label='Cholera cases')
decorate(xlabel='weeks', 
             ylabel='World population (billion)')
```
